**Self Consistency Prompting**

This method runs the same input several times and gives the majority output as the result to increase the accuracy and to reduce bias

**Steps to Perform:**



**Step 1: Set up the OpenAl API Key**



```python
import os
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv('open_ai.env')
client=OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

**Step 2: Define a Function to Get Completion**



```python
%%sql
```

```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0.7
    )
    return response.choices[0].message['content']
```

**Step 3: Define Your Prompts**

```python
prompt = ["""
Let's consider which is heavier: 1000 feathers or a 30-pound weight. I'll think through this in a few different :

1. First line of reasoning: A single feather is very light, almost weightless. So, 1000 feathers might still be quite light.

2. Second line of reasoning: 1000 is a large number, and when you add up the weight of so many feathers, it could be quite heavy.

3. Third line of reasoning: The average weight of a feather is very small. Even 1000 feathers would not add up to 30 pounds.

Considering these reasonings, the most consistent answer is:"""]

response = get_completion(prompt)
print(response)
```

The most consistent answer is that a 30-pound weight is heavier than 1000 feathers.
