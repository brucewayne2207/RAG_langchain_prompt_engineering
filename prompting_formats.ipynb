Zero shot prompting


 it is an type of prompting which can generate the output without any examples and no need to define any output indicator,
 this method uses the priorly trained llm model to generate the output

1.Set up open ai api key

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv('open_ai.env')
client=OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

2.Define a get completion function

```python
def get_completion(prompt,model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0.7
    )
    return response.choices[0].message.content
```

```python
prompt='''
Translate the following text to french:
'Hello world!'
'''
response=get_completion(prompt)
print(response)
```

```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0.7
    )
    return response.choices[0].message.content
```
