**Tree of Thoughts Prompting**

 it is a technique where the model explores multiple distinct reasoning paths a “tree” of candidate thoughts instead of following a single linear chain.
It then evaluates and selects the most promising branches to reach a higher-quality final answer, especially for complex or multi-step problems.


**Steps to Perform:**


**Step 1: Set up the OpenAl API Key**


```python
import os
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv('open_ai.env')
client=OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

**Step 2: Define a Function to Get Completion**


```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0.7
    )
    return response.choices[0].message['content']
```

**Step 3: Define Your Prompts**


```python
prompt = ["""
Solve the problem: A farmer has 100 meters of fencing and wants to enclose the maximum area for his rectangular field. What should the dimens

Let's think about this in a few ways:

1. If the field is a square, each side would be 100 / 4 = 25 meters. The area would be 25 * 25 = 625 square meters.

2. What if the field is not a square? Let's try a 2:1 ratio. The lengths would be 40 and 20 meters. The area would be 40 * 20 = 800.

3. Are there any other ratios that might give a larger area than a square or a 2:1 rectangle?

Considering these options, the best dimensions for the maximum area are:"""]

response = get_completion (prompt)
print("AI Response:")
print(response)
```

AI Response:
The dimensions should be 40 meters by 20 meters, resulting in a maximum area of 800 square meters.
