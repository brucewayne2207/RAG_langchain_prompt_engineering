**Chain of thoughts prompting**

 this is an kind of prompt similar to few shot prompting, where you train your llm model through the input
 so that the model can generate the response in the way you required

**Steps to Perform:**



**Step 1: Set up the OpenAl API Key**



```python
import os
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv('open_ai.env')
client=OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

**Step 2: Define a Function to Get Completion**



```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0.7
    )
    return response.choices[0].message['content']
```

**Step 3: Define Your Prompts**

```python
prompts = ['''
"Imagine you are a detective trying to solve a mystery.",
"You arrive at the crime scene and start looking for clues.",
"You find a strange object at the crime ]scene. What is it?",
"How does this object relate to the crime?",
"Who do you think is the suspect and why?"''']

for prompt in prompts:
       response = get_completion(prompt),
       print(f"Prompt: {prompt}"),
       print(f"Response: {response}"),
       print()
```
